```python
import matplotlib.pyplot as  plt
import numpy as np
import  math

x  = np.linspace(0,25,25)
y = [np.sqrt(i) for i in x]
y2 = x ** 4
plt.xlabel(" X values")
plt.ylabel(" y values")
plt.plot(x,y, "y",label="Square Roots")
plt.plot(x,y2,label="Cube")
plt.title("Graphs of Square roots and Cubes ")
plt.legend(loc ="center left")
```
